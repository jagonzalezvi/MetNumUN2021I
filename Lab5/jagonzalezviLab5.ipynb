```python
class BootCampTask(QCAlgorithm):
    def Initialize(self):
        # 1. Set Starting Cash 
        self.SetCash(1000000)
        self.SetStartDate(2020, 11, 1)
        self.SetEndDate(2021, 4, 1)

        # Select stocks and add in DataNormalizationMode = Raw
        self.adbe = self.AddEquity("ADBE", Resolution.Minute)
        self.adbe.SetDataNormalizationMode(DataNormalizationMode.Raw)
        
        self.t = self.AddEquity("T", Resolution.Minute)
        self.t.SetDataNormalizationMode(DataNormalizationMode.Raw)
        
        self.fb = self.AddEquity("FB", Resolution.Minute)
        self.fb.SetDataNormalizationMode(DataNormalizationMode.Raw)
        
        self.msft = self.AddEquity("MSFT", Resolution.Minute)
        self.msft.SetDataNormalizationMode(DataNormalizationMode.Raw)
        
        self.dell = self.AddEquity("DELL", Resolution.Minute)
        self.dell.SetDataNormalizationMode(DataNormalizationMode.Raw)
        

    def OnData(self, data):

        if not self.Portfolio.Invested:
            self.SetHoldings([PortfolioTarget("ADBE", 0.2), PortfolioTarget("T", 0.2), PortfolioTarget("MSFT", 0.2), PortfolioTarget("FB", 0.2), PortfolioTarget("DELL", 0.2)])
            self.Log("Active Pos Adobe " + str(self.Portfolio["ADBE"].Quantity));
            self.Log("Active Pos At&t " + str(self.Portfolio["T"].Quantity));
            self.Log("Active Pos Microsoft " + str(self.Portfolio["MSFT"].Quantity));
            self.Log("Active Pos Facebook " + str(self.Portfolio["FB"].Quantity));
            self.Log("Active Pos Dell " + str(self.Portfolio["DELL"].Quantity));
```
